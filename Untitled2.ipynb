```python
# ---------------------------------------
# 1. Import Libraries
# ---------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
sns.set_style('Whitegrid')
plt.rcParams['figure.figsize'] = (10,6)

print("Libraries Imported Successfully!")
```

```python
# ---------------------------------------
# 2. Load the Dataset
# ---------------------------------------
# Load the Titanic dataset
data = pd.read_csv('train.csv')  # Replace with your correct file path

# Show first 5 rows
data.head()
```

```python
# ---------------------------------------
# 3. Basic Exploration
# ---------------------------------------
print("Shape of Data:", data.shape)
print("\nData Types and Missing Values:\n")
print(data.info())
print("\nStatistical Summary:\n")
print(data.describe())
print("\nMissing Values:\n")
print(data.isnull().sum())
print("\nUnique Values per Column:\n")
print(data.nunique())
```

```python
# ---------------------------------------
# 4. Univariate Analysis
# ---------------------------------------

# Survived - Categorical
sns.countplot(x='Survived', data=data)
plt.title('Survival Count')
plt.show()

# Pclass - Categorical
sns.countplot(x='Pclass', data=data)
plt.title('Passenger Class Distribution')
plt.show()

# Age - Numerical
sns.histplot(data['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.show()

# Fare - Numerical
sns.boxplot(x=data['Fare'])
plt.title('Fare Boxplot')
plt.show()
```

```python
# ---------------------------------------
# 5. Bivariate Analysis
# ---------------------------------------

# Gender vs Survival
sns.barplot(x='Sex', y='Survived', data=data)
plt.title('Survival Rate by Gender')
plt.show()

# Passenger Class vs Survival
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title('Survival Rate by Class')
plt.show()

# Age vs Fare Scatterplot
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=data)
plt.title('Age vs Fare Colored by Survival')
plt.show()
```

```python
# ---------------------------------------
# 6. Correlation Analysis
# ---------------------------------------

# Heatmap
# Select only numerical features for correlation analysis
numerical_data = data.select_dtypes(include=['number'])
corr = numerical_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Pairplot
sns.pairplot(data.dropna(), hue='Survived', vars=numerical_data.columns) # Specify numerical columns for pairplot
plt.show()
```

```python
# ---------------------------------------
# 7. Summary of Findings
# ---------------------------------------

print("Summary of Key Insights:")

print("""
- Majority of passengers did not survive (around 61% died).
- Females had a much higher survival rate than males.
- Passengers from 1st class had better chances of survival.
- Younger passengers (especially kids) survived more.
- People who paid higher fares had higher survival probability.
- Age has a right-skewed distribution with some missing values.
""")
```
